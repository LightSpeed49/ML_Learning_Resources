```python
import numpy as np
from LinearRegression import LocallyWeightedRegression as LWR
import matplotlib.pyplot as plt
import random
```

```python
x = np.linspace(-3,3,num=50).reshape(1,-1)
x = x.T
y = (x + x**2 + x**3)
```

```python
x.shape, y.shape
```

```python
plt.scatter([i[0] for i in x], y, c="b",s=8)
plt.grid()
plt.title("Training Data y vs x")
plt.show()
```

```python
xq_ls = [np.random.random_sample(size=1)*3*random.sample([1,-1],k=1)[0] for i in range(10)]
y_ls = []

for xq in xq_ls:
    lr = LWR(tow=0.5,epochs=100000, lr=0.1)
    lr.fit(x,y)
    y_ls.append(lr.predict(xq))
```

```python
plt.scatter([i[0] for i in x], y, c="b",s=4)
plt.scatter([i[0] for i in xq_ls], y_ls, c="r",s=4)
plt.title("Blue - Training Points, Red - Query Points")
plt.grid()
```
